```python
from tkinter import *
from PIL import ImageTk,Image

root = Tk()

# width * height
root.geometry('300x300')
root.minsize(200,200)
root.maxsize(1000,1000)
root.title('First Tkinter GUI')

# important Label Options

# text - adds text
# bg = background
# fg - foreground
# padx - x-padding
# pady - y-padding
# relief - border styling - SUNKEN,RAISED,GROOVE,RIDGE

title_label = Label(text=''' my name is varun, we are going out
\n for a hike and then we 
\n will have dinner ''',bg = 'red',fg = 'white',padx = 50,pady=80,font=('comicsansms',19,'bold'),borderwidth=5,relief=SUNKEN)

title_label.pack(side = BOTTOM,anchor='sw',fill=Y,padx=20,pady=20)

## Important Pack Options

# anchor
# side = top,bottom,left.right
# fill
# padx
# pady

root.mainloop()
```

```python
from tkinter import *
import cv2


root = Tk()

root.geometry('400x400')

## function to open Camera



def camera():
    cap = cv2.VideoCapture(0)

    while True:
        ret,frame = cap.read()
        
        cv2.imshow('image',frame)
        if cv2.waitKey(13) and 0XFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

def existing():
    name = input('Enter Name:')
    id = int(input('Enter ID:'))

f1 = Frame(root,bg='grey',borderwidth=4,relief=SUNKEN)
f1.pack(side=TOP)

b1= Button(f1,fg='red',text = 'New Employee',command = camera)
b1.pack(side='left',padx=10)

b2= Button(f1,fg='red',text = 'Existing Employee',command=existing)
b2.pack(side='left',padx=10)

root.mainloop()

```

```python
from tkinter import *



root = Tk()
root.geometry('400x400')

def getVals():
    print(UserValue.get())
    print(PassValue.get())

user = Label(root,text='UserName')
password = Label(root,text = 'Password')

user.grid()
password.grid(row = 1)

### variable classes in tkinter

# BooleanVar, DoubleVar,IntVar,StringVar

UserValue = StringVar()
PassValue = StringVar()

userentry = Entry(root,textvariable=UserValue)
passentry = Entry(root,textvariable=PassValue)

userentry.grid(row=0,column=1)
passentry.grid(row=1,column=1)

Button(text='Submit',command=getVals).grid()

root.mainloop() 
```

```python
## Travel Form using Checkbutons and Entry Widget

from tkinter import *

root = Tk()
root.geometry('400x400')

def getVals():
    print('it works')

def getVals():
    print(namevalue.get())
    print(phonevalue.get())
    print(emailvalue.get())
    print(checkvalue.get())


Label(root,text='Tkinter Basic Program',font = 'comicsansms 13 bold',pady=20).grid(row=0,column=3)

name = Label(root,text='Name')
phone = Label(root,text='phone')
email = Label(root,text='email')

name.grid(row=1,column=2)
phone.grid(row=2,column=2)
email.grid(row=3,column=2)

namevalue = StringVar()
phonevalue = StringVar()
emailvalue = StringVar()
checkvalue = IntVar()


nameentry = Entry(root,textvariable=namevalue)
phoneentry = Entry(root,textvariable=phonevalue)
emailentry = Entry(root,textvariable=emailvalue)

nameentry.grid(row=1,column=3)
phoneentry.grid(row=2,column=3)
emailentry.grid(row=3,column=3)

# checkbox and packing it

check = Checkbutton(text = 'Support Us',variable=checkvalue)
check.grid(row=6,column=3)


Button(text='Submit',command=getVals).grid(row=7,column=3)

root.mainloop() 
```
